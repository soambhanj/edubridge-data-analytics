```python
from tkinter import *
from tkinter import ttk
c=Tk()


c.title(" CURRENCY CONVERTER")
c.geometry("600x400")

# currencies dictionary
currency={
    "American Dollar":75.33,
    "Brazailian Real":14.59,
    "British Pound":100.73,
    "Euro":84.19,
    "Singapore Dollars":55.44,
    "Chinese Yuan":11.94,
    "Japanese Yen":0.65,
    "Canadaian Dollar":59.08
}
# writing code on clicking of button

def con():
    p=inp.get()
    a=currency[v.get()]
    #print(a)
    convert=float(p)*float(a)
    r1.delete(1.0,END)
    r1.insert(INSERT,convert)
    

# putting up name
name=Label(c,text="CURRENCY CONVERTER",font=("Times New Roman",20,"bold","underline"),fg="blue")
name.place(x=150,y=20)

l1=Label(c,text="Enter the Value in Foreign Currency:",font=("Times New Roman",15),fg="Red")
l1.place(x=25,y=70)

inp=Entry(c)
inp.place(x=350,y=70)

l2=Label(c,text="Choose the options:",font=("Times New Roman",15),fg="Red")
l2.place(x=25,y=120)

v=StringVar(c)
v.set(None)
op=OptionMenu(c,v,*currency)
op.place(x=300,y=120,height=50,width=150)
         
l3=Label(c,text="Result in Indian Rupees:",font=("Times New Roman",15),fg="Red")
l3.place(x=25,y=190)     

r1=Text(c,height=2,width=20)
r1.place(x=300,y=190)

b=Button(c, text="Convert",fg="red",font=("Times New Roman",15),command=con)
b.place(x=300,y=250)

c.mainloop()
```
